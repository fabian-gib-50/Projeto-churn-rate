- **`O que é Churn Rate?`** 
     - Churn, numa definição mais generalista, é uma métrica que indica o número de clientes que cancelam em determinado período. Para calcular o churn, o que você precisa fazer é somar o número de clientes que cancelou seu produto/serviço no período analisado.

     - Entender por que seus clientes abandonam o seu produto ou serviço é vital para conquistar um crescimento sustentável. Como o Churn tem um efeito negativo na receita de uma empresa, entender o que é esse indicador e como trabalhar para mitigar essa métrica é algo crítico para o sucesso de muitos negócios.

## Aplicação de procedimentos importantes antes de iníciar outras análises do `conjunto de dados`

  - Aplicar função para **conversão** da variável "TotalCharges"
  - Cópia do Dataset, caso haja algum emprevisto terei uma cópia
  - Eleminação, exclusão da variável `customerID`, não usarei essa variável, otimizando as entradas
  - Converterei a String (object) `TotalCharges` para Float, valor flutuante
  - Preenchimento da variável `TotalCharges` usando método **fillna**, passando para a variável uma `mediana` de valores, para que as variáveis fiquem equivalentes
  - Como observado acima, a quantidade de dados ausentes ( 11 ) fica na casa de 1%, o que não é muito, então podemos preencer com "mediana", esse procedimento não trará nenhm prejuízo para às análises presentes e futuras.

```python
# Função para aplicar a conversão da coluna "TotalCharges" de `String` para `Float64`

def conversao_str_float ( entrada ):
    """
    Converte um objeto do tipo string em float.

    # Arguments
        entrada: string, string da coluna TotalCharges.

    # Returns
        Valor convertido em float, caso permita a conversão.
        NaN, caso o valor não seja apropriado.
    """
    try:
        return float( entrada )
    except ValueError:
        return np.nan
```

```python
# Procedimentos: copiando o database e passando para a variável "df_clean", objeto Python

df_clean = df_raw.copy()

# conversão da string de TotalCharges para float
df_clean[ 'TotalCharges' ] = df_clean[ 'TotalCharges' ].apply( conversao_str_float )

# Depois da conversão ter sido efetuada, agora aparecem ( 11 ) ocorrências de dados ausentes
print()
print( 'Checagem de dados ausentes depois da conversão da variável "TotalCharges":\t[\t{}\t]\n'.format( df_clean[ 'TotalCharges' ].isnull().sum()) )
print( 'Resultado Tipo da variável após conversão:\t', [df_clean.TotalCharges.dtypes ] )

# usar a mediana para substituir os valores ausentes da mesma coluna
TotalChargesMedian = df_clean.TotalCharges.median()
df_clean[ 'TotalCharges' ].fillna(TotalChargesMedian, inplace=True )
```

### Explicação sobre o uso do Método PCA - Área de Análise e Estatístiva `Multi`variada 

 - Análise de Componente Principais – PCA
  - De forma simples, o PCA reduz a dimensionalidade dos dados duplicados, semelhantes ou iguais, e correlações tanto positivas (+) quanto negativas (-), com semelhanças equivalentes.
  - Quais os befefícios: são vários, mas principalmente a redução de ruido nos dados, dimensionalidade e tempo de processamento, além de outras vantagens é claro.
  - O PCA ( do inglês, Principal Component Analysis), é um método matemático usado para maximizar as informações que podem ser extraídas de um conjunto de dados e assim, reduzir o número de variáveis da matriz original dos dados envolvidos no modelamento. Atuando de modo a simplificar a série de dados e reduzir a dimensão dos dados com a finalidade de gerar novas variáveis que não são correlacionadas. Por causa dessa habilidade, diz-se grosso modo que acontece uma “compressão de dados” ( MORGANO; QUEIROZ; FERREIRA, 1999).

  - O PCA é um dos métodos de Reconhecimento de Padrões (PR – Pattern Recognition) Não Supervisionados, os quais são usados para examinar similaridades ou diferenças entre amostras (ROGGO et al., 2007). Outro PR não supervisionado é o HCA (do inglês, Hierarchical Cluster Analysis).

  - A PCA foi introduzida por Karl Pearson em 1901, porém o tratamento formal do método é devido ao trabalho de Hotteling, na década de 30.
  - A análise por PCA é uma técnica exploratória de dados multivariados. Uma das principais vantagens da análise de componentes principais (PCA) é a eliminação de grande parte dos ruídos experimentais. Isto é possível, pois os ruídos constituem uma população de erros aleatórios, ou seja, não estão correlacionados com as informações contidas na matriz de dados. 

```python
# Aplica a redução de dimensionalidade das variáveis do cojunto de dados em questão.

pca = PCA( n_components = 2 ).fit_transform( clust_one )
```

```python
# Determinando um range de K

k_range = range( 1, 10 )

# Aplicando o modelo K-Means para cada valor de K (esta célula pode levar bastante tempo para ser executada)
k_means_var = [KMeans(n_clusters = k).fit(pca) for k in k_range]
```

```python
# Ajustando o centróide do cluster para cada modelo

centroids = [ X.cluster_centers_ for X in k_means_var ]


# Calculando a distância euclidiana de cada ponto de dado para o centróide
k_euclid = [cdist(pca, cent, 'euclidean') for cent in centroids]
dist = [np.min(ke, axis = 1) for ke in k_euclid]
```

```python
# Soma dos quadrados das distâncias dentro do cluster - criação da variável "soma_quadrados_inter_cluster"

soma_quadrados_intra_cluster = [sum( f**2 ) for f in dist ]

# Soma total dos quadrados
soma_total = sum( pdist( pca ) **2) / pca.shape[0]

# Soma dos quadrados entre clusters
soma_quadrados_inter_cluster = soma_total - soma_quadrados_intra_cluster
```

```python
# Plotagem da curva de Elbow - Curva cotovelo

print( )
fig = plt.figure(figsize=( 14, 6) )
ax = fig.add_subplot( 111 )
ax.plot( k_range, soma_quadrados_inter_cluster / soma_total * 100, 'b*-' )
ax.set_ylim(( 0,100 ) ) 
plt.grid( True )
plt.xlabel('Número de Clusters', fontsize=15, color='#003478' )
plt.ylabel( 'Percentual de Variância Explicada', fontsize=15, color='#003478' )
plt.title( 'CURVA DE ELBOW - VARIÂNCIA EXPLICADA ( X ) VALOR  DE ( K ) EM ( y )', pad=15, fontsize=15, color='#003478' )

plt.tight_layout()
```

### Curva Elbow e Mapeamento dos Clusters

   - Como ler e interpretar a curva Elbow e os mapas de Clusters?
   
     - No eixo_X temos o número de Clusters que foram criados
     - Eixo_Y temos o Percentil da Variãncia Explicada.
       - Percentil de `Variãncia Explicada`: quanto mais consigo explicar a variância dos dados, maior e melhor serão as divisões dos Clusteres, correpondente ao Eixo_X.
     - Ex: Cluster de número ( 3 ), estou conseguindo explicar ( 90% ) da variância dos dados, quanto mais próximo de ( 100% ) melhor, mas se observar o platô do Cluster 9 ao 10, significa que as variâncias são menos explicadas, na prática a curva vai se perdendo, acima do Cluster 7, logo crio o modelo com no máximo 8 Clusteres.
     - Obs.: Não é a quantidade de Clusters que irá melhorar o percentil da variância, quanto mais subdivisão, menos explica-se as variãncias e isso interfere nos resultados finais, nos indicadores de divisão e classificação dos clusters.
     
   - Centroids
     - Ex: Os centroides, ou centro dos clusters, são marcações de posição central da aglutinação de cada cluster, quanto mais concentrado melhor, significa que o modelo implementado, e as métricas estão performando. Denota também quão concentrado está cada segmentação, cluster.
     
  - Silhouette Score 
    - O indicador é de 0.7808, isso significa que o modelo criado com ( 6 ) Clusteres, têve uma acurácia de 78%, entre 70 e 90% é considerado ótimo.
    - Para esse conjunto de dados com ( 8 ) Clusteres, o indicador ficou ficou em torno de 65%, significa bom, mas com 6 ficou ótimo.
    - No agrupamento, chamo a variável K **modelo_v1** para trazer somente os dados referente ao processo com ( 6 ): Clusters, que foi o melhor indicador gerado, com acurácia de 78%.
  - Na prática com a Curva Elbow, conseguimos ter um ótimo direcionamento do número de Clusteres que devemos aplicar.
  
  - Farei aplicação dos processos com dados padronizados, acreditando que as métricas ficarão ainda melhores.

### Validação de Hipóteses 
   - Criação de uma Função Multipla com elementos e modelagem da área de Estatisticas
     - Temos alguns métodos que são utilizados sistemáticamente em projetos de **Data Scientist**, principalmente os médotos de visualização, mas não são únicos.        Aqui na função podemos dizer que existem dois métodos, o **Inferêncial e o descritivo**.
     - Farei um código para testar e validar algumas `Hipóteses`. Acho que vale a pena investigar ainda mais, esse assunto tem tudo haver com métodos e metodologias que eu defendo e me especializei nos últimos anos, `"Inteligência de Mercado Competitiva"`.
   - Farei na sequência com a função criada, várias chamadas para validar proporções das variáveis, confrontando  com cancelamentos "Exit".
      - Acho pertinente nesse momento, fazer algumas levantadas de hipóteses para as ocorrências de cancelamentos e sua relação com as outras variáveis.
      - Do ponto de vista Inteligência de Mercado Competitiva, será crucial que surjam dúvidas dos times de negocios, principalmente retenção, relacional, suporte e área financeira.
      - Como diria o grande Físico **"Albert Einstein"**: `"Não podemos esperar resultados direntes, se tudo que fazemos é repetir sempre as mesmas coisas".`

```python
# Criei uma "Função Multipla", para validar hipóteses e enteder melhor as proporções dos cancelamentos (prop1, prop2)
# Coeficientes usasos e modelagem, são conceitos área da estatística "inferêncial"

# Função para cáculos de proporções 
def proporcao_churn ( df_cleanx,lista,z_base ):
    
  # Calculando amostragem 1:
  churn_1 = df_cleanx[ 'Yes' ] [0]
  n_1     = df_cleanx.sum( axis=1 )[0]
  prop_1  = n_1/df_clean.shape[0]

  # Calculando amostragem 2:
  churn_2 = df_cleanx[ 'Yes' ] [1]
  n_2     = df_cleanx.sum( axis=1 )[1]
  prop_2  = n_2/df_clean.shape[0]

  # Criação de DataFrame com os dados estatísticos
  df_hip  = pd.DataFrame( [[ churn_1,n_1,prop_1 ],
                           [ churn_2,n_2,prop_2 ]],
                           index = lista,
                           columns = ['X','n','proporções'] )
  print(  df_hip )
  print( '\n' )
  p_bar = ( prop_1*n_1+prop_2*n_2 ) / df_clean.shape[0]
  z = ( prop_1-prop_2 ) / np.sqrt( p_bar*( 1-p_bar )*( 1/n_1+1/n_2 ) )
  if np.absolute(z) > z_base:
    print( 'Se o p-valor < alpha, H0 então rejeite!' )
  elif np.absolute(z) < z_base:
    print( 'Se o p-valor > alpha, imprima H0.' )
    
# Função proporção e médias
def mean( df_cleanx,lista ):
  # Normalizando os calculos da amostragem 1:
  churn_1 = normalize( df_cleanx,norm='l1' )[0,1]
  n_1 = df_cleanx.sum(axis=1)[0]
  sigma_1 = np.sqrt(churn_1*(1-churn_1))

  # Normalizando os calculos da amostragem 2:
  churn_2 = normalize( df_cleanx,norm='l1' )[1,1]
  n_2 = df_cleanx.sum(axis=1)[1]
  sigma_2 = np.sqrt(churn_2*(1-churn_2) )

  # Criando DataFrame com dados estatísticos
  df_hip = pd.DataFrame( [[ churn_1,n_1,sigma_1 ],
                            [ churn_2,n_2,sigma_2 ]],
                            index = lista,
                            columns = [ 'X_bar','n','sigma' ] )
  print(  df_hip )
  print( '\n' )
  z = ( churn_1-churn_2 ) / np.sqrt((( sigma_1**2 ) / n_1)+(( sigma_2**2 ) /n_2 ) )
  print( 'O z-score encontrato é: {};'.format(z) )
```
